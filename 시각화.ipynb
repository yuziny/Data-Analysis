# 시각화

```python
import seaborn as sns
import pandas as pd
import plotly.express as px

df = sns.load_dataset('taxis', cache=False)
df.head()
```

```python
df['pickup_date'] = df['pickup'].dt.date
df.head()
```

```python
df_agg1 = df.groupby(by=['payment','pickup_date']).total.mean().reset_index()
df_agg1.head()
```

```python
import plotly.express as px


#선 그래프
line_fig = px.line(df_agg1
                   ,x='pickup_date'
                   ,y='total'
                   ,color='payment')
line_fig.show()
```

```python
df_agg2 = df.groupby(by=['payment','pickup_date','color']).total.mean().reset_index()
df_agg2.head()
```

```python
import plotly.express as px


#선 그래프
line_fig = px.line(df_agg2
                   ,x='pickup_date'
                   ,y='total'
                   ,color='payment'
                   ,facet_row='color' #서브플롯을 구분하기 위한 컬럼(세로로 나누기)
                   )
line_fig.show()
```

# histogram

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
import plotly.express as px

fig = px.histogram(
    iris,
    x='sepal_length',
    color='species',
    barmode='overlay', # group/overlay/stack 중 선택
    marginal='rug', # 추가적인 데이터 포인트 분포 표시
    title='histogram',
    nbins=30 # 막대 수
)

fig.show()
```

```python
import plotly.express as px

fig = px.histogram(
    iris,
    x='sepal_length',
    color='species',
    barmode='overlay', # group/overlay/stack 중 선택
    marginal='rug', # 추가적인 데이터 포인트 분포 표시
    title='histogram',
    nbins=30 # 막대 수
)
# 세로선 추가
fig.add_vline(
    x=6,
)
# 가로선 추가
fig.add_hline(
    y=6,
)
# 주석 추가
fig.add_annotation(
    x=6,
    y=15,
    text='텍스트'
)
# 세로 스팬
fig.add_vrect(
    x0='5.5',
    x1='6.5',
    fillcolor='red',
    opacity=0.2, # 스팬의 불투명도
)
# 가로 스팬
fig.add_hrect(
    y0='10',
    y1='15',
    fillcolor='green',
    opacity=0.2, # 스팬의 불투명도
)
fig.show()
```

```python
iris_name = 'setosa'
filtered_iris = iris[iris.species == iris_name]
filtered_iris.species.unique()
```

```python
avg = filtered_iris.sepal_length.mean()
std = filtered_iris.sepal_length.std()

fig = px.histogram(
    filtered_iris,
    x='sepal_length',
    height=500,
    width=700,
    title=f'{iris_name} histogram'
)
fig.add_vline(
    x=avg,
)
fig.add_annotation(
    x=avg,
    y=30,
    text='{:.2f}'.format(avg)
)
fig.add_vrect(
    x0=avg-std,
    x1=avg+std,
    fillcolor='red',
    opacity=.2,
)
fig.show()
```

```python
array1 = iris.species.unique()
for iris_name in array1:
    filtered_iris = iris[iris.species == iris_name]
    avg = filtered_iris.sepal_length.mean()
    std = filtered_iris.sepal_length.std()

    fig = px.histogram(
        filtered_iris,
        x='sepal_length',
        height=300,
        width=500,
        title=f'{iris_name} histogram',
        nbins=20
    )
    fig.update_xaxes(range=[3,10])
    fig.update_yaxes(range=[0,12])
    fig.add_vline(
        x=avg,
    )
    fig.add_annotation(
        x=avg,
        y=30,
        text='{:.2f}'.format(avg)
    )
    fig.add_vrect(
        x0=avg-std,
        x1=avg+std,
        fillcolor='red',
        opacity=.2,
    )
    fig.show()
```
