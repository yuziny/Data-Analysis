```python
# 파일의 세 번째 줄(인덱스로는 2)에 헤더(열 이름)가 있다고 지정
seoul_business = pd.read_csv('/content/drive/MyDrive/포트폴리오/4. business_type.csv', sep=',', header=2)
seoul_business
```

```python
seoul_business = seoul_business.iloc[:, [0,1,3,5,13,17,19,21]]
```

```python
seoul_business
```

```python
seoul_business = seoul_business.loc[seoul_business['동'] == '소계']
seoul_business.rename(columns={'종사자수.1':'한식일반음식점업 종사자수',
                               '종사자수':'택시운송업',
                               '종사자수.5':'커피전문점 종사자수',
                               '종사자수.7':'기타주점업 종사자수',
                               '종사자수.8':'일반교과학원 종사자수',
                               '종사자수.9':'한식육류요리전문점 종사자수'},
                      inplace=True)
seoul_business
```

```python
seoul_business.to_csv('df_seoul_business.csv', index=False)
```
